```python
sus = 0
err = 0
for i in range(0,len(mov)):
    try:
        mov["IMDB ID"][i]=data[data["title"]==mov["Movie Title"][i]]["titleId"].unique()[0]
        sus = sus+1
    except:
        
        err = err +1
    print("Succcess Rate is  :",round((sus/len(mov))*100,2))
    print("Error    Rate is  :",round((err/len(mov))*100,2))
    print("Completation rate :",round((i/len(mov))*100,2))
    clear_output(wait=True)   
```
