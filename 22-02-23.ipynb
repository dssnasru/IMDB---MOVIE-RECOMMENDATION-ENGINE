```python
import requests
from bs4 import BeautifulSoup as bs
HEADERS = {'User-Agent': 'Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148'}
def movie_scrapper(id):
    url = f"https://www.imdb.com/title/{id}/"
    page=requests.get(url,headers=HEADERS)
    data=bs(page.content)
    movie ={
        "Title":data.find(class_="cEmnhL").text,
        "Release":data.find(class_="hMnkBf").text,
        "Rating":data.find(class_="eUYAaq").text,
        "Genre": [x.text for x in data.find_all(class_="ipc-chip__text") ][0:-1],
        "Cast":[x.text for x in data.find_all(class_="sc-bfec09a1-1 fUguci") ]
    }
    return movie
```

```python
movie_scrapper("tt2380307")
```

```python
mov = ["tt1049413","tt0382932","tt0435761","tt2380307"]
```

```python
emt_mov = []
for i in mov:
    x = movie_scrapper(i)
    emt_mov.append(x)
    
```
